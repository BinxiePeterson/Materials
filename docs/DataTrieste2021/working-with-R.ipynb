## Working with Jupyter notebook and R

Download datasets (e.g. temperature) from the Climate Change Knowledge Portal and calculate and plot the average monthly temperature using Jupyter notebook and R

1 Visit: https://climateknowledgeportal.worldbank.org/download-data 

2 Select ‘Temperature’, the country and the time period you are interested

3 Click ‘Download Data’, and save the CSV file on your computer as Temperatures.csv


4 Go to the training instance of the Jupyter Notebook service: 
http://Your_IP_Address:8888

5 Wait for your Jupyter notebook to boot up

6 Upload Temperatures.csv


7 Open a new R Notebook and save it with a new name


8 Installing gdata() and gtools() and press “Run”

```r
install.packages("gtools")
install.packages("gdata")
```

9 Use the gdata() library and the read.csv() method to read the CSV file and press “Run”

```r
library(gdata)
temp <- read.csv(file="Temperatures.csv", header=TRUE, sep=",")
```

10 Use the head() method to display the first few rows of the imported dataset (press “Run”)

```r
head(temp)
```

11 Use the aggregate() function to group temperatures per Year, and to calculate the mean for each year (press “Run”)

```r
datasets = aggregate(temp[, 1:2], list(temp$Year), mean)
```

12 Print the average mean temperature per year (press “Run”)

```r
print(datasets)
```

13 Plot the graph (press “Run”)

```r
plot (datasets[3:2], type="l", col="blue", main="Average Temperature per Year", xlab="Year", ylab="Temperature (°C)")
```
