7 Plot the dataset

The scatterplot shows data from the dataset `mammals` (MASS) and it is an extreme version of the common situation where positive (or non-zero) values are squashed together in the lower part of the range, with a tail out to the right. Such a distribution is said to be “skewed to the right”.

```r
plot(brain ~ body , data=mammals)
mtext(side=3, line=0.5, adj=0, "A: Unlogged data")
```

Next plot in Figure shows the scatterplot for the logged data.

```r
plot(brain ~ body , data=mammals, log="xy")
mtext(side=3, line=0.5, adj=0,
"B: Log scales (both axes)")
```
